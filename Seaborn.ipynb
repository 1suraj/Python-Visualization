# Matrix plot

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
sns.barplot(x='month',y='passengers',data=flights)
```

```python
heatdata=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(heatdata,cmap='magma',linecolor='black',linewidth=1)
```

```python
sns.clustermap(heatdata,cmap='coolwarm',standard_scale=1)
```

# Regression_Plot

```python
sns.lmplot(x='tip',y='total_bill',data=tips,hue='sex',markers=['o','+'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='tip',y='total_bill',data=tips,col='sex',row='time',aspect=0.5,size=8)
```
