```python
import numpy as np
```

```python
x=np.linspace(0,10,20)
y=x**2
```

#functional Method

```python
plt.plot(x,y,'g--')
#plt.show()
#xlabel1=str(Numbers)
#ylabel1=str(Squred Numbers)
plt.xlabel('Numbers')
plt.ylabel('Squred Numbers')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g--')
plt.subplot(1,2,2)
plt.plot(y,x,'r--')
```

object Oriented

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes2.plot(x,-y,'b')

axes1.set_ylabel('sq numbers')
axes2.set_ylabel('sq numbers')

axes1.set_xlabel('numbers')
axes2.set_xlabel('numbers')

axes1.set_title('my chart')
axes2.set_title('my small chart')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.tight_layout()

# axes are the array in a list so we can llop through them

#for myaxes in axes:
#   myaxes.plot(x,y)
#Also there are list so we can do these too.

axes[0].plot(x,y,'y')
axes[1].plot(y,x,'b--')
```

# Size and DPI of viz

```python
fig=plt.figure(figsize=(8,3))

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes2.plot(x,-y,'b')

```

# Multiple Line:

```python
fig=plt.figure(figsize=(8,8))

axes=fig.add_axes([0.1,0.1,0.8,0.8])


axes.plot(x,y**3,'r',label='X3')
axes.plot(x,y**2.5,'g',label='X2.5')

axes.legend(loc=10)
#axes.legend(loc=(0.1,0.1))
```

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,color='green',linewidth=1,alpha=1,linestyle='-',marker='*',markersize=25,markerfacecolor='yellow',markeredgewidth=3
       markeredgecolor='red')
#linewidth=lw
#linewidth=ls
```

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,color='green')
```
